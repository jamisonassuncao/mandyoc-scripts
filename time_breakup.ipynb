```python
import os
import numpy as np
import matplotlib.pyplot as plt

path = os.getcwd().split('/')
machine_path = '/'+path[1]+'/'+path[2] #cat the /home/user/ or /Users/user from system using path

##File of fancy plot style. To use des.mplstyle file you must have an LaTeX distribution in your system
plt.style.use(machine_path+'/Mestrado/cenarios/mandyoc/scripts_gera_inputs_imgs/new_outputs/des.mplstyle')

%matplotlib inline
```

```python
Tps80 = [1292, 1350, 1397]
tb_lit80_C1 = [19.11, 24.66, 27.14]
tb_lit80_C10 = [19.0, 24.74, 26.91]

Tps150 = [1352, 1421, 1460]
tb_lit150_C1 = [16.72, 16.47, 17.59]
tb_lit150_C10 = [19.91, 18.93, 19.08]

fig, ax = plt.subplots(1, 1, figsize=(8,6), constrained_layout=True)

#Plot data
ax.plot(Tps80, tb_lit80_C1, 'o', markersize=12, alpha=1.0, color = 'xkcd:cerulean blue')
ax.plot(Tps80, tb_lit80_C10, 'o', markersize=12, alpha=1.0, color = 'xkcd:scarlet')

ax.plot(Tps80, tb_lit80_C1, '-', alpha=0.5, color = 'xkcd:cerulean blue')
ax.plot(Tps80, tb_lit80_C10, '-', alpha=0.5, color = 'xkcd:scarlet')


ax.plot(Tps150, tb_lit150_C1, '^', markersize=12, alpha=1.0, color = 'xkcd:cerulean blue')
ax.plot(Tps150, tb_lit150_C10, '^', markersize=12, alpha=1.0, color = 'xkcd:scarlet')

ax.plot(Tps150, tb_lit150_C1, '-', alpha=0.5, color = 'xkcd:cerulean blue')
ax.plot(Tps150, tb_lit150_C10, '-', alpha=0.5, color = 'xkcd:scarlet')

#Plot texts
for i in range(len(Tps80)):
    ax.text(Tps80[i], tb_lit80_C10[i]+0.5, f'{Tps80[i]}$^\circ$C', color='black', fontsize=16)
    ax.text(Tps150[i]-7, tb_lit150_C10[i]+0.5, f'{Tps150[i]}$^\circ$C', color='black', fontsize=16)

#ghost points
ax.plot(-10, -10, 'o', color='black', label = r'$H_{\mathrm{lit}}$ = 80 km')
ax.plot(-10, -10, '^', color='black', label = r'$H_{\mathrm{lit}}$ = 150 km')
ax.legend(loc='upper right', handletextpad=0.15)

ax1 = ax.twinx()
ax1.axis('off')
ax1.plot(-10, -10, '-', color = 'xkcd:cerulean blue', label = 'Weak lower crust')
ax1.plot(-10, -10, '-', color = 'xkcd:scarlet', label = 'Strong lower crust')
ax1.legend(loc='upper left', labelcolor='linecolor', handlelength=0, handletextpad=0)

ax.set_xlim([1290, 1470])
ax.set_ylim([16, 30])

ax.set_xlabel('Tp [$^\circ$C]')
ax.set_ylabel('Time of breakup [Myr]')
ax.grid('-k', alpha=0.6)
```
